```python
import earthaccess
import os

# 1. Authenticate (Interactive - enter credentials when asked)
auth = earthaccess.login(strategy="interactive", persist=True)

# 2. Define the Manzanillo "Bounding Box"
# Format: [min_lon, min_lat, max_lon, max_lat]
bbox = (-104.45, 19.00, -104.20, 19.20)

# 3. Search NASA's database for SRTM data
results = earthaccess.search_data(
    short_name="SRTMGL1",
    bounding_box=bbox
)

# 4. Define download path (Relative to where this notebook is saved)
# We go "up" one level (..), then into data/raw/srtm
download_path = "../data/raw/srtm"

# 5. Download
print(f"Found {len(results)} files. Downloading...")
earthaccess.download(results, download_path)
print("Download Complete!")
```
